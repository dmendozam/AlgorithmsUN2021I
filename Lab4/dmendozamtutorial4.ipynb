#Numpy and Basic Pandas.

NumPy and Pandas are some of the stronger and more useful python libraries. A library is a package of functions and objects one can import to it's work in order to make things easier and make use of what other people have achieved.

```python
#So here is how we import a library.

import numpy as np

import pandas as pd

# the 'as np' and 'as pd' work as aliases for the libraries.
```

* In this tutorial we'll se NumPy's arrays and Panda's series.

```python
# A Numpy's array is like a python list but:

# It is size fixed.
# All the data must be the same type

list_numbers=[1,2,3,4,5]
array_numbers=np.array(list_numbers)

print (list_numbers, 'type: ', type(list_numbers))
print (array_numbers, 'type: ', type(array_numbers))
```

It's well known that numpy's arrays are better for mathematical operations than python's lists.

```python
#A Numpy's array can be multidimensional too.
multidimension_array = np.array([[True,True], [False,True]])
print(multidimension_array)

#We can even see the shape of the array (matrix)
print(multidimension_array.shape)

#For accesing rows we use indexes
print(multidimension_array[0])

#For accesing columns we  use the following
print(multidimension_array[:,0])
```

There are some special functions we can apply to an numpy's array in order to modify the entire array or produce a single value.



```python
#some function examples
print(np.log(array_numbers))

print(np.mean(array_numbers))
```

By pandas we have series. These are one dimensional arrays which can hold any type of objects. It is special because it's labeled and it performs specially well when dealing with financial data. 

```python
price = [143.73, 145.83, 143.68, 144.02, 143.5, 142.62]

s = pd.Series(price, index = ['a', 'b', 'c', 'd', 'e', 'f'])

print(s)

#Series can be indexed as Python's list or as dictionaries.
print("indexaciones: ")
print(s[2:4])
print(s["b"])

#We can give the series a name too.
s.name="prueba"
print(s)

```

Series have functions too. For example we have the describe function which gives us some statistical data of the series.

```python
print(s.describe())
```

There is an special funcion within pandas, it is calle date_range and we use it to crete time indexes. Time indexes are ase regular indexes but we can specify the date format and the frequency of the dates.

```python
time_index = pd.date_range('2017-01-01', periods = len(s), freq = 'D')
print (time_index)
s.index = time_index
print (s)
```
