* Functions as reusable blocks of code.

```python
def sum(x,y):
  return x+y
```

The above block shows a basic funtion, where we use the 'def' as a keyword for python to know we're aboyt to write a function. The items in between the parentheses are the function's arguments or the items we need to introduce in the function in order to it for work.

```python
sum_1=sum(2,3)
sum_2=sum(5,7)
sum_3=sum(sum_1,sum_2)
print(sum_1, sum_2, sum_3)
```

There are some functions that are already defined in the python programming language. These are called 'Built-In Functions', these functions are thinked as solutions for frequent programming problems to give the programmer a faster work flow.

Some examples of these built-in fuctions are shown below.

```python
#range() is a function that creates a list containing an arithmetic sequence. 
#The arguments must be integers.
#The default step in between the numbers is 1.

print (range(5))
print (*range(5))
```

```python
#The function len() returns as one might expect the lenght of an object.

print(len(range(5)))
```

```python
#Map is a function that applies a specific function to every item of
# a sequence or collection.

print(map(len,['123','12345']))  #if we don't specify that we want the results as a list
print(list(map(len,['123','12345'])))
```

```python
#Lambda functions are small anonymous functions used only where they are created

list(map(lambda x: x*2, range(5)))
```

* A class is a type of data, just like a list. When we create an object based on that class we are instancing the class.

* In python, everything is an object.

* We can create our own objects by defining a class.



```python
class Car:
    #in the __init__ function we declare our class attributes.
    def __init__(self, color, kind, year, brand):
        self.color = color
        self.kind = kind
        self.year = year
        self.brand = brand

    def change_color(self, new_color):  #This is a class function.
        self.color = new_color
```

Now that we have created our class we can create an object based on it. We can access its attributes and functions.

```python
red_car= Car('red','suv','2021','Mercedes Benz') #Creation of an object

print('The current color is: ',red_car.color) #Acccesing object's attributes.

red_car.change_color('blue') #Accesing object's functions.

print('The new color is: ',red_car.color, 'It works!')
```

* When we create a class that has all the attributes and methods of a parent class we are doing something known as 'Inheritance'.

```python
class SuperCar(Car):
  def __init__(self, horsePower):
    self.horsePower=horsePower

ferrari= SuperCar(1000)

print(ferrari.horsePower)

ferrari.change_color('yellow')  #SuperCar has Car functions.

print(ferrari.color)  #SuperCar has Car attributes.
```
